```python
import numpy as np

# Assuming 'df' is your DataFrame with standardized data
# 'columns_to_standardize' contains the column names you want to handle

# Define the z-score threshold for outlier removal (e.g., threshold = 3)
z_score_threshold = 0.7

# Create a mask to identify outliers for each specified column
outlier_masks = {}
for column in columns_to_standardize:
    z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())
    outlier_masks[column] = z_scores <= z_score_threshold

# Create an overall outlier mask (logical OR of individual column masks)
overall_outlier_mask = np.logical_and.reduce(list(outlier_masks.values()))

# Apply the overall outlier mask to filter out rows with outliers
outlier_filtered_df = df[~overall_outlier_mask]

# Display the shape of the filtered DataFrame to see how many outliers were removed
print(f"Original shape: {df.shape}")
print(f"Shape after removing outliers: {outlier_filtered_df.shape}")
```
